Q1. Install Docker and create Dockerfiles to containerize applications?


Below are Steps to Install Docker on Ubuntu

```python
sudo apt update
sudo apt install docker.io -y
sudo systemctl start docker
sudo systemctl enable docker
```

To Verfiy whether it is Installed or not

```python
docker --version
docker run hello-world
```

I Created Small Sample Application to prepare Dockerfile

```python
#app.js

const express = require('express');
const app = express();

app.get('/', (req, res) => {
    res.send('Docker Assignment Working!');
});

app.listen(3000, () => {
    console.log('Server running on port 3000');
});
```

```python
#package.json

{
  "name": "docker-assignment",
  "version": "1.0.0",
  "main": "app.js",
  "dependencies": {
    "express": "^4.18.2"
  }
}
```

```python
#DockerFile

# Use official Node base image
FROM node:18

# Set working directory
WORKDIR /app

# Copy package files
COPY package*.json ./

# Install dependencies
RUN npm install

# Copy app source
COPY . .

# Expose port
EXPOSE 3000

# Start application
CMD ["node", "app.js"]
```

Build Docker Image

```python
docker build -t docker-assignment .
```

Check image:

```python
docker images
```

Run Container

```python
docker run -p 3000:3000 docker-assignment
```

open browser local to your system

```python
http://localhost:3000
```
