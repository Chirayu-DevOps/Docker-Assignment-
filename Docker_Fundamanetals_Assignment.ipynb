Q1. Use Docker Compose to define and run multi-container applications?

Step 1: Install Docker Compose

```python
docker compose version
#if not installed check belwo command

sudo apt install docker-compose-plugin -y
```

```python
#app/app.js

const express = require('express');
const app = express();

app.get('/', (req, res) => {
    res.send('Docker Compose Multi-container App Working');
});

app.listen(3000, () => {
    console.log('App running on port 3000');
});
```

```python
#app/package.json
{
  "name": "compose-app",
  "version": "1.0.0",
  "dependencies": {
    "express": "^4.18.2"
  }
}
```

```python
#DockerFile
FROM node:18-alpine

WORKDIR /app

COPY package*.json ./

RUN npm install

COPY . .

EXPOSE 3000

CMD ["node", "app.js"]
```

```python
#docker-compose.yml

version: '3.8'

services:

  app:
    build: ./app
    container_name: node_app
    ports:
      - "3000:3000"
    depends_on:
      - db

  db:
    image: mysql:8
    container_name: mysql_db
    environment:
      MYSQL_ROOT_PASSWORD: root
      MYSQL_DATABASE: testdb
    ports:
      - "3306:3306"
```

Step 6: Start Containers

Run from project root:

docker compose up -d

```python
#Check running containers:

docker ps
```
